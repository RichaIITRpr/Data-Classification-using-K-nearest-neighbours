## KNN Classifier
Using KNN Classifier to classify given dataset.
Carrying out KNN on two types of datasets
        
        1. Standrdized (or Z-Score Normalization)
        2. Normalized (or Min-Max Scaling)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
def knn_classifier(x_train, x_test, y_train, y_test, iters_range):
    error_rate = []
    accuracy = []
    for i in range(1, iters_range, 2):
        knn = KNeighborsClassifier(n_neighbors=i)
        knn.fit(x_train, y_train)
        prediction = knn.predict(x_test)
        
        # Printing the Accuracies and Confusion matrix for each K
        print('K = {:}'.format(i))
        print('Accuracy: {:.2f}'.format(accuracy_score(prediction,y_test)*100))
        print('Confusion Matrix :\n',confusion_matrix(y_test, prediction),'\n')
#         print('Classification Report:\n',classification_report(y_test,prediction))
        print('\n-----------------------------------------------------------\n')
        
        error_rate.append(np.mean(prediction != y_test))
        accuracy.append(accuracy_score(prediction,y_test)*100)
    return error_rate, accuracy
```

# 1. Standardizing the variables


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df.drop('Class', axis =1))
```

```python
scaled_features_std = scaler.transform(df.drop('Class',axis=1))
```

```python
df_features = scaler.transform(df.drop('Class',axis=1))
df_features = pd.DataFrame(scaled_features_std, columns = df.columns[:-1])
df_features.head()
```

```python
x_train_std, x_test_std, y_train_std, y_test_std = train_test_split(scaled_features_std,df['Class'], test_size=0.30)
```

```python
iters_range = 40
```

```python
different_knn_standardized, accuracy_standardized = knn_classifier(x_train_std, x_test_std, y_train_std, y_test_std, iters_range)
```

```python
print('Maximum accuracy achieveed: ',max(accuracy_standardized))
```

```python
print('Minimum error ', min(different_knn_standardized),' at ', different_knn_standardized.index(min(different_knn_standardized)))
```

```python
plt.figure(figsize = (10,6))
plt.plot(range(1,iters_range,2) , different_knn_standardized, color='blue',linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10 )
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
print('From the above plot and from the accuracy values:\n\tK =', different_knn_standardized.index(min(different_knn_standardized))+1, '\ngives the least error with \n\taccuracy = ',max(accuracy_standardized), '%.\nHence, K =',different_knn_standardized.index(min(different_knn_standardized))+1, ' is our required K-value.')
```
